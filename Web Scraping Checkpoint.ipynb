```python
#Question 1.1
import requests as re
from bs4 import BeautifulSoup
r= re.get("https://en.wikipedia.org/wiki/The_Hobbit")
print(r.text)
soup=BeautifulSoup(r.content,'html.parser')
print(soup.prettify())
```

```python
#Question 1.2
print(soup.title)
```

```python
#Question 1.3
print(soup.get_text())
```

```python
#Question 1.4
for link in soup.find_all('a'):
    print(link.get('href'))
```
